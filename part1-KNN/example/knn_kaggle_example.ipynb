```python
def type_to_numeric(x):
        if x=='white': return 0
        if x=='red':   return 1
```

```python
from sklearn import preprocessing

#wine = wine.dropna(how='all')
wine.fillna(0, inplace=True)


y = wine['type'].apply(type_to_numeric)
#y = y.values.reshape(-1,1)

#X = wine[['chlorides']]
wine.drop(['type'],axis='columns',inplace=True)
X = wine


X

```

```python
# Import train_test_split function
from sklearn.model_selection import train_test_split

# Split dataset into training set and test set
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3) # 70% training and 30% test
```

```python
#Import knearest neighbors Classifier model
from sklearn.neighbors import KNeighborsClassifier

#Create KNN Classifier
knn = KNeighborsClassifier(n_neighbors=5)

#Train the model using the training sets
knn.fit(X_train, y_train)

#Predict the response for test dataset
y_pred = knn.predict(X_test)
```

```python
knn.score(X_test, y_test)
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import cross_val_score
scores = cross_val_score(KNeighborsClassifier(n_neighbors=5), X, y,cv=3)
np.average(scores)
```
